# Optimized OOD Validation Pipeline

This notebook validates the trained model on 2024+ OOD data using saved artifacts. 
- Loads fine-tuned MiniLM for embeddings.
- Applies saved scaler and LR model for predictions.
- Computes classification and financial metrics.
- Optimized for speed with batching and minimal overhead.

## Prerequisites
- Artifacts: 'artifacts/numeric_scaler.joblib', 'artifacts/lr_model.joblib'
- Local MiniLM: './minilm_finphrasebank_final_model/'
- OOD Data: 'combined_data_with_lags.csv' (2024+ with headlines, prices, etc.)


# Relevant Imports

```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, 
    roc_auc_score, matthews_corrcoef, confusion_matrix, ConfusionMatrixDisplay
)
import joblib
import pathlib
import json
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print("Using device:", device)

# Helper function for Sharpe Ratio
def calc_sharpe(ret_series, annualiser=252):
    mean_ret = ret_series.mean()
    std_ret = ret_series.std()
    return (mean_ret / std_ret) * np.sqrt(annualiser) if std_ret > 0 else 0.0
```

# Configuration

```python
# Paths
DATA_FILE = pathlib.Path('combined_data_with_lags.csv')  # OOD data (2024+)
ARTIFACT_DIR = pathlib.Path('artifacts')
SCALER_FILE = ARTIFACT_DIR / 'numeric_scaler.joblib'
LR_FILE = ARTIFACT_DIR / 'lr_model.joblib'
LOCAL_MODEL = pathlib.Path('./minilm_finphrasebank_final_model')
BASE_MODEL = "microsoft/MiniLM-L12-H384-uncased"  # For tokenizer

# Model params
MAX_LENGTH = 512  # Matches MiniLM default
BATCH_SIZE = 32   # For efficient embedding generation
NUMERICAL_COLS = [
    'Open', 'sentiment_volatility_lag1', 'aggregate_sentiment_score_lag1',
    'Close_lag1', 'High_lag1', 'Volume_lag1', 'Daily_Return_lag1', 'Volatility_lag1'
]  # From training (8 cols)

# Output dir
RESULTS_DIR = pathlib.Path('results')
RESULTS_DIR.mkdir(exist_ok=True)
```

# Model Architecture

```python
# Load scaler (expects 392 features)
try:
    scaler = joblib.load(SCALER_FILE)
    if scaler.n_features_in_ != 392:
        raise ValueError("✗ Loaded scaler expects {} features, but should be 392.".format(scaler.n_features_in_))
    print("✓ Loaded 392-dim scaler")
except FileNotFoundError:
    raise FileNotFoundError(f"✗ {SCALER_FILE} not found. Run training notebook first.")

# Load LR model
try:
    lr_model = joblib.load(LR_FILE)
    print("✓ Loaded LR model")
except FileNotFoundError:
    raise FileNotFoundError(f"✗ {LR_FILE} not found. Run training notebook first.")

# Load tokenizer (base) and fine-tuned MiniLM model
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
text_model = AutoModel.from_pretrained(LOCAL_MODEL)
text_model.to(device)
text_model.eval()
print("✓ Loaded text model and tokenizer")
```

```python
# Load data
ood_data = pd.read_csv(DATA_FILE)
print(f"✓ Loaded {len(ood_data)} rows from {DATA_FILE}")

# Convert Date to datetime and sort
ood_data['Date'] = pd.to_datetime(ood_data['Date'])
ood_data = ood_data.sort_values('Date').reset_index(drop=True)

# Validate required columns
required_cols = ['Date', 'title', 'content'] + NUMERICAL_COLS
missing_cols = [col for col in required_cols if col not in ood_data.columns]
if missing_cols:
    raise ValueError(f"✗ Missing columns: {missing_cols}")

# Drop any unnecessary rolling/GARCH features (as in training)
rolling_features = [
    'garch_cond_variance_lag1', 'garch_cond_volatility_lag1', 'garch_residuals_lag1',
    'rolling_cond_volatility_3_lag1', 'rolling_cond_volatility_5_lag1'
]
ood_data = ood_data.drop(columns=[col for col in rolling_features if col in ood_data.columns], errors='ignore')
print("✓ Data loaded and validated")
print(ood_data.head())
```

```python
# Aggregate raw numerics daily (mean) - use RAW (unscaled) to build concat
num_daily_raw = ood_data[NUMERICAL_COLS + ['Date']].groupby('Date').mean().reset_index()

# Merge with shifted embeddings
merged = pd.merge(num_daily_raw, daily_emb_shifted, on='Date', how='inner')

# Concatenate raw num + text embeds for full market vector (392 dim)
text_cols = [c for c in merged.columns if c.startswith('e')]
text_mat = merged[text_cols].values  # Shape: (n_days, 384)
num_mat_raw = merged[NUMERICAL_COLS].values  # RAW num: (n_days, 8)
X_concat = np.hstack([num_mat_raw, text_mat])  # (n_days, 392)

# Now scale the full concatenated vector with loaded scaler
X_scaled = scaler.transform(X_concat)
print("✓ Built and scaled full 392-dim features")
```

```python
# Predict with LR model on scaled features
y_prob = lr_model.predict_proba(X_scaled)[:, 1]
y_hat = (y_prob > 0.05).astype(int)

df_preds = pd.DataFrame({'Date': merged['Date'], 'y_prob': y_prob, 'y_hat': y_hat})
print("✓ Generated predictions")
print(df_preds.head())
```

```python
# Create targets from prices (unique daily)
prices = ood_data[['Date', 'Close', 'Daily_Return']].drop_duplicates().sort_values('Date')
prices['ret_1d'] = prices['Close'].pct_change().shift(-1)
prices['target'] = (prices['ret_1d'] > 0).astype(int)

# Merge predictions with targets
merged_preds = pd.merge(df_preds, prices[['Date', 'target', 'ret_1d']], on='Date', how='inner')

# Classification metrics
acc = accuracy_score(merged_preds['target'], merged_preds['y_hat'])
precision = precision_score(merged_preds['target'], merged_preds['y_hat'])
recall = recall_score(merged_preds['target'], merged_preds['y_hat'])
f1 = f1_score(merged_preds['target'], merged_preds['y_hat'])
roc_auc = roc_auc_score(merged_preds['target'], merged_preds['y_prob'])
mcc = matthews_corrcoef(merged_preds['target'], merged_preds['y_hat'])

print(f"Accuracy: {acc:.3f} | Precision: {precision:.3f} | Recall: {recall:.3f}")
print(f"F1: {f1:.3f} | ROC-AUC: {roc_auc:.3f} | MCC: {mcc:.3f}")

# Confusion matrix
cm = confusion_matrix(merged_preds['target'], merged_preds['y_hat'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Down (0)", "Up (1)"])
disp.plot(cmap="Blues")
plt.title("OOD Confusion Matrix")
plt.savefig(RESULTS_DIR / 'ood_confusion_matrix.png')
plt.show()

# Financial metrics
merged_preds['strategy_ret'] = merged_preds['y_hat'] * merged_preds['ret_1d']
sharpe = calc_sharpe(merged_preds['strategy_ret'])
win_rate = np.sum(merged_preds['strategy_ret'] > 0) / len(merged_preds) if len(merged_preds) > 0 else 0
winning_sum = np.sum(merged_preds['strategy_ret'][merged_preds['strategy_ret'] > 0])
losing_sum = np.sum(merged_preds['strategy_ret'][merged_preds['strategy_ret'] < 0])
profit_factor = winning_sum / abs(losing_sum) if abs(losing_sum) > 1e-10 else np.nan

print(f"Sharpe Ratio: {sharpe:.2f} | Win Rate: {win_rate:.2f} | Profit Factor: {profit_factor:.2f}")
```
